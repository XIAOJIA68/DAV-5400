# Question1

```python
import pandas as pd
import numpy as np

# read the data file from GitHub and convert to a NumPy ndarray object

df = pd.read_csv("https://raw.githubusercontent.com/XIAOJIA68/DAV-5400/master/Assn10.%20Tidying%20data.csv")
df
```

# Question2

```python
# drop the blank row which does not have any meaning and is unregular in this data set 

df = df.drop(2)
df
```

```python
# add names for the first two columns, the airline is for the first column and the time_delay is for the second column
# rename the columns

df.columns = ["airline","time_delay","Los Angeles","Phoenix","San Diego","San Francisco","Seattle"]
df
```

```python
# Convert the table to Long format using airline & time_delay as the keys and Los Angeles,Phoenix,San Diego,San Francisco 
# and Seattle as the data values

df = df.melt(id_vars = ["airline","time_delay"], value_vars = ['Los Angeles','Phoenix','San Diego','San Francisco','Seattle'])
df
```

```python
# Correctly fill the NaN items in the 'airline' column
# Based on the content of the original data frame, it appears that the 'NaN' values in the 'airline' column should be 
# filled with the 'airline' value from the preceding row. We can use the fillna() function to do this:

df['airline'].fillna(method='ffill', inplace = True)
df
```

# 3.Q(2) Which airline had the best overall on time performance?

```python
# for analysis of on time performance, at first i need to get the data which time_delay column equals on time 
# get the data which time_delay column equals on time 
df_ontime = df[df["time_delay"] == 'on time']
df_ontime
```

```python
# to analyze different airlines, we should group by each airline and get their overall on time value.
# store the overall result to a new datafram

df_ontime =df_ontime.groupby('airline').sum()
df_ontime
```

from the result above, we can see that the AM WEST has the best overall on time performance

# Q4: transform your “long” data back into a “wide” format

i would mimic the structure of the graphic shown above to transform my “long” data back into a “wide” format. At first, i would drop the blank row as i did before because it's unmeaningful, and then i will change the airline to be the index instead of filling the blank value, finally i will use pivot to transfer the values in variable column to be single city columns. I think this wide format is already easier and more understandable for views.

```python
# use pivot to transfer the values in variable column to be different single city column

wide = pd.pivot_table(df, values = 'value', index = ['airline',"time_delay"],columns = ['variable'])
wide
```
