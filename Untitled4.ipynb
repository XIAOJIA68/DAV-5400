```python
# import csv
# define numpy as np
# define panda as pd
# print the result as df
import csv 
import numpy as np
import pandas as pd
with open (r'F:\Analytics Programming\Project-3\tournamentinfo.txt') as txtfile:
    readCSV = csv.reader(txtfile)
    table = []
    for row in readCSV:
        table.append(row)
table
```

```python
#remove the first three unmeaningful rows, the rows after the forth can seem like a regular formulas
table1 = table[3:]
table1
```

```python
#use np to get the rows which contain the name of player
#create a list to store rows

length = len(table1)

b = np.arange(1,length,3)
row = 0
table2 = []
for a in table1:
    if (row in b):
        table2.append(a)
    row = row+1
table2
```

```python
#use np to get all rows which contain states 
#create a list to store rows

length = len(table1)
c = np.arange(2,length,3)
row = 0
table3 = []
for a in table1:
    if (row in c):
        table3.append(a)
    row = row+1
table3
```

```python
#import re to extract the name

import re
str1 = re.compile('[A-Z]+\s[A-Z]+\s[A-Z]*')
name = str1.findall(str(table2))
name
```

```python
#import re to extract the state

str2 = re.compile('[A-Z]{2}')
state = str2.findall(str(table3))
state
```

```python
#import re to extract the total Number of Points

str3 = re.compile('[0-9]\.[0-9]')
rate = str3.findall(str(table2))
rate
```

```python
#import re to extract the player's number 

str12 = re.compile('\s[0-9]{1,2}\s')
num1 = str12.findall(str(table2))
num1
num = re.compile('[0-9]+')
num = num.findall(str(num1))
num
```

```python
#import re to extract the opponents' number

str10 = re.compile('[A-Z]{1}\s+[0-9]{1,2}')
opponents10 = str10.findall(str(table2))
opponents10

str11 = re.compile('[0-9]+')
opponents11 = str11.findall(str(opponents10))
opponents11
```

```python
#import re to extract the Player’s Pre-Rating

str4 = re.compile('[A-Z]{1}\:\s+[0-9]{3,4}')
pre1 = str4.findall(str(table3))
pre1

str5 = re.compile('[0-9]+')
pre2 = str5.findall(str(pre1))
pre2
```

```python
#create the datafram to store the information of each player 

data = {'Num':num,'Name':name,'State':state,'Pre_Rating':rate,'Points':pre2} 
# Create DataFrame 
df = pd.DataFrame(data) 
df
```

```python
#export Pandas DataFrame to a CSV File

df.to_csv(r'F:\Analytics Programming\Project-3\Project3.csv')
```
