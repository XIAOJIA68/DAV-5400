```python
import pandas as pd
import numpy as np

# read the data file from GitHub and convert to a NumPy ndarray object
filename = "https://raw.githubusercontent.com/jtopor/DAV-5400/master/Project1/hflights.csv" 
hflights = pd.read_csv(filename)
hflights
```

narrative for transformations:
to get the average delay-time, at first we need to sum the total delay of ArrDelay and DepDelay.and then we should use groupby to category the flights and get the mean.Finally we use max to get the maximum average delay time. 

To get the total dealy time of the flights, we create a new column called totaltime_delay to store the sum of ArrDelay and DepDelay

```python
#create a new column called totaltime_delay to store the sum of ArrDelay and DepDelay

hflights['Totaltime_delay'] = hflights['ArrDelay'] + hflights['DepDelay']

hflights
```

```python
#after get the total delay time, we should group by each flight and get their average delay time by different TailNum
#show the new datafram

ave_TailNum = hflights.groupby('TailNum').mean()
ave_TailNum
```
